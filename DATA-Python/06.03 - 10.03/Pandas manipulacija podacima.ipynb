Zašto nam je bitno imati odgovarajuće tipove podataka za svaki stupac?

Možemo pokazati na primjeru `last_review` stupca.

Idemo dodati novi stupac u naš cijeli `dataframe` u kojem će biti zapisana samo godina zadnjeg reviewa.

```python
listings_df["last_review_year"] = listings_df["last_review"].dt.year
listings_df
```

Nakon dodavanja novih stupaca i manipulacija nad podacima u postojećim stupcima, obično je dobro provjeriti da li nam se nešto
promijenilo (slučajno ili namjerno) s tipovima podataka.

```python
listings_df.dtypes

listings_df["last_review_year"] = listings_df["last_review_year"].astype("int64")
```

Nakon što smo zadovoljni tipovima podataka s kojima radimo, možemo se pobliže upoznati sa samim podacima.

Jedna od osnovnih stvari prilikom čišćenje bilo kojeg seta podataka je vidjeti imamo li prazne vrijednosti i trebamo li nešto napraviti s njima.

```python
listings_df.info()
```

```python
listings_df.isnull()  # vraća True tamo gdje imamo vrijednost, a False tamo gdje je nemamo
```

```python
listings_df.isnull().sum()
```

Sad znamo u kojim stupcima imamo praznih vrijednosti.

Ovisno o analizi koji želimo napraviti, te prazne vrijednosti nam mogu smetati i prije obrade podataka ih želimo "srediti".

U ovom konkretnom slučaju želimo prazne vrijednosti u stupcu:
- `name` postaviti na `No name listing`
- `host_name` postaviti na `J. Doe`
- `last_review` postaviti na `01-01-1970`
- `last_review_year` postaviti na `0`
- `reviews_per_month` postaviti na `0.00`


Zamjena praznih vrijednosti se najčešće radi pomoću metodu `fillna`.

```python
listings_df["name"] = listings_df["name"].fillna("No name listing")
```

```python
listings_df["host_name"] = listings_df["host_name"].fillna("J. Doe")
```

```python
listings_df.isnull().sum()
```
