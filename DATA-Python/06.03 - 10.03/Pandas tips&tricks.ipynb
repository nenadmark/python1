## 6. Iteriranje po `dateframe-u`

```python
for i, row in cars.iterrows():
    if row["Origin"] == "US":
        # .loc[i] -> locira redak s tim indeksom
        # .loc[i, "<j>"] -> locira vrijednost u retku "i" te stupcu "j"
        cars.loc[i, "American_made"] = True   
    else:
        cars.loc[i, "American_made"] = False
```

```python
cars
```

```python
del cars["American_made"]
cars.head()
```

```python
# puno bolji način za ovo bi bio pomoću vektorizacija

cars["American_made"] = cars["Origin"] == "US"
cars
del cars["American_made"]
```

```python
cars["American_made"] = cars.apply(lambda row: row["Origin"] == "US", axis=1)
cars.head()
```

## 8. Kreiranje više `dataframe-ova` u međukoracima

```python
cars_hp = cars.query("Horsepower > 200")
cars_hp
```

```python
cars_model_grouped = cars_hp.groupby(["Model"])["Horsepower"].mean()
cars_model_grouped
```

```python
cars_avg_hp_sorted = cars_model_grouped.sort_values()
cars_avg_hp_sorted
```

```python
cars_avg_hp_sorted_chain = cars.query("Horsepower > 200").groupby(["Model"])["Horsepower"].mean().sort_values()
cars_avg_hp_sorted_chain
```

## 9. "Ručno" grupiranje

```python
osobe = pd.DataFrame(
    {
        "Ime": ["Pero", "Ana", "Marko", "Josipa"],
        "Prezime": ["Peric", "Anic", "Markic", "Josic"],
        "Spol": ["M", "F", "M", "F"],
        "Datum_rodenja": ["01.01.2000", "02.02.2005", "03.03.2010", "04.04.2015"],
        "Punoljetan": ["DA", "DA", "NE", "NE"],
        "Rezultat": [7, 9, 4, 6]
    }
)
```

```python
osobe
```

```python
prosjek_m = osobe.loc[osobe["Spol"] == "M"]["Rezultat"].mean()
prosjek_f = osobe.loc[osobe["Spol"] == "F"]["Rezultat"].mean()

print(prosjek_m)
print(prosjek_f)
```

```python
prosjeci = osobe.groupby(["Spol"])["Rezultat"].mean()
prosjeci
```

```python
podaci_po_spolu = osobe.groupby(["Spol"])["Rezultat"].agg(["min", "max", "mean"])
podaci_po_spolu
```
