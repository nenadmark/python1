## 1. - Happiness score (WHR) 

- This is the national average response to the question of life evaluations asking the following: â€œPlease imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you.
- This measure is also referred to as Cantril life ladder.
- Gapminder has converted this indicator's scale from 0 to 100 to easly communicate it in terms of percentage.


- Norway: 7. place

```python
df_happiness = pd.read_csv("hapiscore_whr.csv")
df_happiness.set_index("country", inplace = True)
df_happiness.head()
```

```python
top10by_happiness = df_happiness["2022"].sort_values(ascending=False).head(10)
top10by_happiness
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.bar(top10by_happiness.index, top10by_happiness)
```
