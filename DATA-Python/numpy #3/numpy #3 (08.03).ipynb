Ovo nije dajsretniji format iz kojeg mozemo nesto zakljuciti jer nam se sve nalazi u jednom stupcu.

zato sto su stupci u nasem .csv file-u odvojeni s ";" dok pandas kao defaultni separator ocekuje ","

Ovaj problem mozemo vrlo lako rijesiti

```python
cars = pd.read_csv("podaci/cars.csv", sep=";")
```

```python
cars = cars.drop(0)
# micanje redka jer prvi redak je string
# da bi pretvorili horsepower stupac iz double u float, 
#moramo prvi redak cijeli maknuti

cars["Horsepower"] = cars["Horsepower"].astype("float")
cars
```

```python
cars.dtypes
```

```python
cars["Car"] = cars["Car"].astype("string")
cars.dtypes
```

```python
cars["MPG"] = cars["MPG"].astype("float")
cars.dtypes
```

```python
cars["Cylinders"] = cars["Cylinders"].astype("int")
cars.dtypes
```

```python
cars["Displacement"] = cars["Displacement"].astype("float")
cars.dtypes
```

```python
cars["Weight"] = cars["Weight"].astype("float")
cars.dtypes
```

```python
cars["Acceleration"] = cars["Acceleration"].astype("float")
cars.dtypes
```

```python
cars["Model"] = cars["Model"].astype("int")
cars.dtypes
```

```python
cars["Origin"] = cars["Origin"].astype("string")
cars.dtypes
```
