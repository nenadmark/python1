```python
zadatak_lista = [6, 3, 1, 10, 5, 7, 9, 2, 2, 2]
```

```python
lista_even = [i for i in zadatak_lista if i % 2 == 0]
lista_odd = [i for i in zadatak_lista if i % 1 == 0]
for i in lista_even:
    print(i)
```

 ista stvar s numpy:

```python
np_originalna = np.array(zadatak_lista)
np_indeksi_parnih = np.where(np_originalna % 2 == 0)
np_indeksi_neparnih = np.where(np_originalna % 1 == 0)
```

```python
print(np_indeksi_parnih)
print(np_indeksi_neparnih)
```

```python
np_originalna = np.array(zadatak_lista)
np_lista_parnih = np_originalna[np_originalna % 2 == 0]
np_lista_neparnih = np_originalna[np_originalna % 2 != 0]
```

```python
print(np_lista_parnih)
print(np_lista_neparnih)
```

### Zadatak

napisati program koji generira nasumicne brojeve od 1 do 100. broijeve je potrebno generirati dok god se svaki broj ne generira barem jednom. na kraju ispisati koliko je puta svaki broj bio generiran

```python
# rjesenje bez numpy-a

import time

brojevi = [0] * 10000

start_time = time.time()
while 0 in brojevi: 
    broj = rd.randint(1, 10000)
    brojevi[broj - 1] += 1
end_time = time.time()
print(f"Ukupno je generirano {sum(brojevi)} brojeva u {end_time - start_time} sekundi.")   
for idx, broj in enumerate(brojevi):
    print(f"Broj {idx + 1} se pojavio {broj} puta.")
```

```python
# rjesenje s numpy-em

brojevi_np = np.zeros(10000, dtype=np.int32)
numbers_generated = set()
total_count = 0

start_time = time.time()

while len(numbers_generated) < 10000:
    num = np.random.randint(1, 10001)
    brojevi_np[num-1] += 1
    numbers_generated.add(num)
    
end_time = time.time()
print(f"Ukupno je generirano {np.sum(brojevi_np)} brojeva u {end_time - start_time} sekundi.")
```
