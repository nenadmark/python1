```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

orange = "#B45400"
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
df.info()
```

```python
sort_attack = df["Attack"].sort_values()


fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.bar(sort_attack.index, sort_attack)
```

```python
ukupno = len(df)
groupby_legendary = df.groupby(["Legendary"])["Legendary"].count()
postotci_legendary = round(groupby_legendary / ukupno, 2)
postotci_legendary
```

```python
fig, ax = plt.subplots()
ax.pie(postotci_legendary.values, labels=['Not Legendary', 'Legendary'], autopct='%1.1f%%')
plt.title("Legendary pokemon percentage", fontfamily="serif", fontsize=15, loc="left",
         y=1.2)
plt.show()
```

```python
groupby_generation = df.groupby(["Generation"])["Generation"].count()
postotci_generation = round(groupby_generation / ukupno, 2)


fig, ax = plt.subplots()
ax.pie(postotci_generation.values, labels=postotci_generation.index, autopct='%1.1f%%')
plt.title("Pokemon generation percentage", fontfamily="serif", fontsize=15, loc="left",
         y=1.2)

plt.show()
```

```python
# Type 1
groupby_type1 = df.groupby(["Type 1"])["Type 1"].count().sort_values(ascending=False)



fig, ax = plt.subplots(1, 1, figsize=(10, 6),
                       layout='constrained', dpi=100, 
                       label='Pokemon Types')
ax.bar(groupby_type1.index, groupby_type1, color=orange)
plt.show()
```
