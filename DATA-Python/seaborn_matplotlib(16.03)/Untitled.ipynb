### Matplotlib

```python
plt.rcParams['figure.figsize'] = [10, 8] 
plt.rcdefaults()
```

```python
okvir = plt.figure(figsize=[8, 8])
graf_1 = okvir.add_axes([0, 2, 0.5, 0.5])
graf_2 = okvir.add_axes([1, 2, 0.5, 0.5])

graf_1.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":m")
graf_2.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":m")
```

```python
okvir_a = plt.figure(figsize=[8, 8])
grid = okvir_a.add_gridspec(3, 3)
graf_gore_lijevo = okvir_a.add_subplot(grid[0, 0])
graf_sredina = okvir_a.add_subplot(grid[1, 1])
graf_dolje_desno = okvir_a.add_subplot(grid[2, :]) #slicing sluzi tu za stretchanje grafa


graf_gore_lijevo.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":m")
graf_sredina.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":m")
graf_dolje_desno.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":m")
```

```python
okvir_b = plt.figure(figsize=[8, 8])
grid_b = okvir_b.add_gridspec(3, 3)

graf_b_1 = okvir_b.add_subplot(grid[0, 0])
graf_b_2 = okvir_b.add_subplot(grid[1, 0])
graf_b_3 = okvir_b.add_subplot(grid[0:2, 1:3]) #slicing sluzi tu za stretchanje grafa
graf_b_4 = okvir_b.add_subplot(grid[2, 1:])


graf_b_1.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":r")
graf_b_2.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":r")
graf_b_3.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":r")
graf_b_4.plot(prosjek_po_datumima.dateRep, prosjek_po_datumima.cases, ":r")
```

```python
import matplotlib.dates as mdates
```

```python
cases_df["dateRep"] = mdates.date2num(cases_df["dateRep"])
```

```python
cases_df["dateRep"]
```

```python
cases_df.plot()
```

```python
cases_df.plot(
    kind="line",  # tip grafa
    x="dateRep",
    y="cases",
    figsize=(10, 5),
    title="Broj COVID slucajeva"
)
```

```python
cases_df.groupby("dateRep", as_index=False).mean().plot(
    kind="line",  # tip grafa
    x="dateRep",
    y=["cases", "deaths"],
    figsize=(10, 5),
    title="Broj COVID slucajeva",
    color=["green", "red"],
    linestyle="solid"
)
```

```python
cases_df.groupby("dateRep", as_index=False).mean().plot(
    kind="line",  # tip grafa
    x="dateRep",
    y=["cases", "deaths"],
    figsize=(10, 5),
    title="Broj COVID slucajeva",
    color=["green", "red"],
    subplots=True,
    linestyle="solid"
)
```

```python
cases_df.groupby("dateRep", as_index=False).mean().plot(
    kind="line",  # tip grafa
    x="dateRep",
    y=["cases", "deaths"],
    figsize=(10, 5),
    title="Broj COVID slucajeva",
    color=["green", "red"],
    subplots=True, #subplots dolazi iz axisa
    layout=(1, 2),
    linestyle="solid"
)
```
